# Integraçao dos dados

### Imports

```python
import pandas as pd
import numpy as np
```

## Carregando raw data

### Dados dos municípios

```python
df_municipio = pd.read_csv('data/municipio/municipios_IBGE_ANSI.txt', sep=",", encoding='latin1')
df_municipio.head()
```

```python
df_populacao = pd.read_csv('data/municipio/populacaoMunicipiosEstimada2011_2021.csv', sep=";", skiprows=4, nrows=5597, header=None)
df_populacao.drop(columns=[4 ,6], inplace=True)
df_populacao.columns = ["tipo", "codigo_ibge", "nome", "populacao_2011", "populacao_2021"]
df_populacao.head()
```

```python
df_area = pd.read_excel(open('data/municipio/AreaMunicipios_2021.xls', 'rb'), sheet_name='AR_BR_MUN_2021', nrows=5572) 
df_area.head()
```

### Dados da Região de Saúde

```python
df_rs_municipio = pd.read_csv('data/regiao_saude_municipio.csv', sep=";")
df_rs_municipio.head()
```

```python
df_regiao_saude = pd.read_csv('data/regiao_saude_geo.csv', sep=",")
df_regiao_saude.columns = names=["cod_reg_saude", "nome_reg_saude", "latitude", "longitude", "limite_geografico"]
df_regiao_saude.head()
```

### Dados Rede Sentinela

```python
df_serv_sentinela = pd.read_csv('data/servicos_sentinelas.csv', sep=",")
df_serv_sentinela.head()
```

### Dados Estados

```python
df_area_estado = pd.read_excel(open('data/municipio/AreaMunicipios_2021.xls', 'rb'), sheet_name='AR_BR_UF_2021', nrows=27) 
df_area_estado.head()

```

### Dados Fluxo Aéreo

```python
df_fluxo_aereo = pd.read_excel(open('data/fluxo_transporte/LIG_AEREAS_2019-2020_fluxos_od.xlsx', 'rb'), sheet_name='BASE_FLUXO_OD', 
                               dtype = {'ID_LIG': int, 'COD_CID_O': int, 'COD_CID_D': int}) 
df_fluxo_aereo.head()
```

### Dados Fluxo Rodoviário e Hidroviário

```python
df_rod_hid = pd.read_csv('data/fluxo_transporte/lig_rodo_hidro.csv', sep=",")
df_rod_hid.head()
```

## Criando tabelas do banco

### Tabela Município

Atributos:
* cod_mun (chave primária) ok
* cod_uf (chave estrangeira) ok
* cod_reg_saude (chave estrangeira) --> feito merge com 1 dígito a menos
* nome_mun ok
* latitude ok
* longitude ok
* populacao_2011 ok
* populacao_2021 ok
* area ok
* densidade_2021
* nome_uf
* uf
* pais

```python
tb_municipio = df_municipio[["codigo_ibge", "nome", "latitude", "longitude", "codigo_uf"]]
tb_municipio.head()
```

```python
tb_municipio = pd.merge(tb_municipio,df_populacao[['codigo_ibge','populacao_2011', 'populacao_2021']],on='codigo_ibge', how='left')
tb_municipio.head()
```

```python
tb_municipio = tb_municipio.merge(df_area[['CD_MUN','AR_MUN_2021']], how='left', left_on='codigo_ibge', right_on='CD_MUN').drop(columns='CD_MUN')
tb_municipio.rename(columns={"AR_MUN_2021": "area", "codigo_uf":"cod_uf", "nome":"nome_mun"}, inplace=True)
tb_municipio.head()
```

```python
tb_municipio['densidade_2021']= tb_municipio['populacao_2021']/tb_municipio['area']
tb_municipio.head()
```

```python
#Necessario fazer merge com a coluna com o digito a menos, pois os dados 
#da região de saúde estão faltando 1 dígito
tb_municipio["col_apoio"] = tb_municipio['codigo_ibge'].astype(str).str[:-1].astype(int)
tb_municipio = tb_municipio.merge(df_rs_municipio[['ibge','regiao']], how='left', left_on='col_apoio', right_on='ibge').drop(columns=['ibge', 'col_apoio'])
tb_municipio.rename(columns={"regiao": "cod_reg_saude", "codigo_ibge":"cod_mun"}, inplace=True)
tb_municipio.head()
```

```python
tb_municipio['populacao_2011'] = tb_municipio['populacao_2011'].replace(to_replace ='...',value = np.nan)
tb_municipio['populacao_2011'] = tb_municipio['populacao_2011'].astype(int, errors='ignore') 
```

Fazendo merge com os dados de estado

```python
tb_municipio = tb_municipio.merge(df_area_estado[['CD_UF','NM_UF', 'NM_UF_SIGLA']], how='left', left_on='cod_uf', right_on='CD_UF').drop(columns=['CD_UF'])
tb_municipio.rename(columns={"NM_UF": "nome_uf", "NM_UF_SIGLA":"uf"}, inplace=True)
tb_municipio.head()
```

Adicionando o país Brasil nas cidades -> Será criado outra dataframe para as cidades não brasileira

```python
tb_municipio['pais'] = "BRA"
tb_municipio.head()
```

Dataframe com cidades não brasileira 
* Retirada dos dados de fluxo rodoviario e hidroviario
* Encontrado somente nas cidades B

```python
# COD_UF_A	UF_A	CODMUNDV_A
df_mun_not_brasil = df_rod_hid[df_rod_hid["UF_B"].str.len()>2][["COD_UF_B","UF_B","CODMUNDV_B", "NOMEMUN_B"]]
df_mun_not_brasil.rename(columns={"COD_UF_B":"cod_uf", "UF_B":"pais", "NOMEMUN_B":"nome_mun", "CODMUNDV_B":"cod_mun"}, inplace=True)
df_mun_not_brasil.drop_duplicates(subset=['cod_mun'], inplace=True)
df_mun_not_brasil.head(10)
```

Concatenando os dataframes com as cidades brasileira e não brasileiras

```python
tb_municipio = pd.concat([tb_municipio,df_mun_not_brasil], axis=0, ignore_index=True)
tb_municipio.tail()
```

### Tabela de Fluxo Aéreo

Atributos:
* id_lig ok (chave primária)
* total_passagens_2019 ok
* total_passagens_2020 ok
* tempo_medio_2019 ok
* tempo_medio_2020 ok
* cod_mun_origem ok (chave estrangeira)
* cod_mun_destino ok (chave estrangeira)

```python
tb_fluxo_aereo = pd.DataFrame(df_fluxo_aereo[['COD_CID_O', 'COD_CID_D', 'VAR01', 'VAR02', 'VAR05', 'VAR06']])
tb_fluxo_aereo.rename(columns={"ID_LIG": "id_lig", "COD_CID_O": "cod_mun_origem", "COD_CID_D": "cod_mun_destino", 
                          "VAR01":"total_passagens_2019", "VAR02":"total_passagens_2020", "VAR05":"tempo_medio_2019",
                          "VAR06":"tempo_medio_2020"}, inplace=True)

# tb_fluxo_aereo = tb_fluxo_aereo.fillna(-1)
tb_fluxo_aereo['total_passagens_2019'] = tb_fluxo_aereo['total_passagens_2019'].astype(int, errors='ignore')
tb_fluxo_aereo['total_passagens_2020'] = tb_fluxo_aereo['total_passagens_2020'].astype(int, errors='ignore')

tb_fluxo_aereo = tb_fluxo_aereo.replace('Não há voo direto','-2', regex=True)
tb_fluxo_aereo['tempo_medio_2019'] = tb_fluxo_aereo['tempo_medio_2019'].astype(int, errors='ignore')
tb_fluxo_aereo['tempo_medio_2020'] = tb_fluxo_aereo['tempo_medio_2020'].astype(int, errors='ignore')

tb_fluxo_aereo.head()
```
