```python
import folium
import json
import pandas as pd

# Carregue seu GeoJSON dos estados do Brasil
estados_json = json.load(open('./br-states.json'))

print(estados_json)

# Criar um DataFrame de exemplo (substitua isso pelos seus dados reais)
data = {
    "Estado": ["Acre", "Amazonas", "Bahia"],  # Substitua pelos nomes dos estados
    "Dado": [25, 42, 18]  # Substitua pelos dados numéricos correspondentes
}

df = pd.DataFrame(data)

# Crie o mapa
mapa = folium.Map(location=[-15.842866, -48.026985], 
                  width=600,
                  height=500,
                  zoom_start=5)

# Use o Choropleth para colorir os estados com base nos dados do DataFrame
folium.Choropleth(
    geo_data=estados_json,
    data=df,
    columns=["Estado", "Dado"],  # Colunas do DataFrame contendo os estados e dados
    key_on="feature.properties.nome",  # Substitua pelo campo de nome no GeoJSON
    fill_color="YlGnBu",  # Escolha uma paleta de cores
    legend_name="Exemplo de Choropleth"  # Nome da legenda
).add_to(mapa)

# Exibir o mapa
mapa
```
