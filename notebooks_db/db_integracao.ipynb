## Criando tabelas do banco

### Tabela Município

Atributos:
* cod_mun (chave primária) ok
* cod_uf (chave estrangeira) ok

```python
tb_municipio = df_municipio[["codigo_ibge", "nome", "latitude", "longitude", "codigo_uf"]]
tb_municipio.head()
```

```python
tb_municipio = pd.merge(tb_municipio,df_populacao[['codigo_ibge','populacao_2011', 'populacao_2021']],on='codigo_ibge', how='left')
tb_municipio.head()
```

```python
tb_municipio = tb_municipio.merge(df_area[['CD_MUN','AR_MUN_2021']], how='left', left_on='codigo_ibge', right_on='CD_MUN').drop(columns='CD_MUN')
tb_municipio.rename(columns={"AR_MUN_2021": "area", "codigo_uf":"cod_uf", "nome":"nome_mun"}, inplace=True)
tb_municipio.head()
```

```python
tb_municipio['densidade_2021']= tb_municipio['populacao_2021']/tb_municipio['area']
tb_municipio.head()
```

```python
#Necessario fazer merge com a coluna com o digito a menos, pois os dados 
#da região de saúde estão faltando 1 dígito
tb_municipio["col_apoio"] = tb_municipio['codigo_ibge'].astype(str).str[:-1].astype(int)
tb_municipio = tb_municipio.merge(df_rs_municipio[['ibge','regiao']], how='left', left_on='col_apoio', right_on='ibge').drop(columns=['ibge', 'col_apoio'])
tb_municipio.rename(columns={"regiao": "cod_reg_saude", "codigo_ibge":"cod_mun"}, inplace=True)
tb_municipio.head()
```

```python
tb_municipio['populacao_2011'] = tb_municipio['populacao_2011'].replace(to_replace ='...',value = np.nan)
tb_municipio['populacao_2011'] = tb_municipio['populacao_2011'].astype(int, errors='ignore') 
```

Fazendo merge com os dados de estado

```python
tb_municipio = tb_municipio.merge(df_area_estado[['CD_UF','NM_UF', 'NM_UF_SIGLA']], how='left', left_on='cod_uf', right_on='CD_UF').drop(columns=['CD_UF'])
tb_municipio.rename(columns={"NM_UF": "nome_uf", "NM_UF_SIGLA":"uf"}, inplace=True)
tb_municipio.head()
```

Adicionando o país Brasil nas cidades -> Será criado outra dataframe para as cidades não brasileira

```python
tb_municipio['pais'] = "BRA"
tb_municipio.head()
```

Dataframe com cidades não brasileira 
* Retirada dos dados de fluxo rodoviario e hidroviario
* Encontrado somente nas cidades B

```python
# COD_UF_A	UF_A	CODMUNDV_A
df_mun_not_brasil = df_rod_hid[df_rod_hid["UF_B"].str.len()>2][["COD_UF_B","UF_B","CODMUNDV_B", "NOMEMUN_B"]]
df_mun_not_brasil.rename(columns={"COD_UF_B":"cod_uf", "UF_B":"pais", "NOMEMUN_B":"nome_mun", "CODMUNDV_B":"cod_mun"}, inplace=True)
df_mun_not_brasil.drop_duplicates(subset=['cod_mun'], inplace=True)
df_mun_not_brasil.head(10)
```

Concatenando os dataframes com as cidades brasileira e não brasileiras

```python
tb_municipio = pd.concat([tb_municipio,df_mun_not_brasil], axis=0, ignore_index=True)
tb_municipio.tail()
```

### Tabela de dados do REGIC

Tabela com diverso atributos para a verificacao da correlacao com o número de casos e mortes de COVID

```python
tb_regic = pd.DataFrame(df_regic_cidades[["COD_CIDADE", "VAR03", "VAR04", "VAR05", "VAR06","VAR07",
                        "VAR09","VAR19","VAR20","VAR21","VAR56", "VAR57", "VAR58", "VAR59", "VAR60", 
                        "VAR61", "VAR62", "VAR63", "VAR64", "VAR65", "VAR66", "VAR95" ]])

tb_regic.rename(columns={"COD_CIDADE":"cod_mun", "VAR03": "PIB", "VAR04": "PIB_agro", "VAR05": "PIB_ind","VAR06": "PIB_serv", "VAR07": "PIB_pub", 
                        "VAR09": "hierarquia","VAR19":"CGT", "VAR20":"CI", "VAR21":"CGP", 
                        "VAR56":"indice_atracao", "VAR57":"ia_vestuario","VAR58":"ia_moveis","VAR59":"ia_saude_bm", "VAR60":"ia_saude_a",
                        "VAR61":"ia_ensino_sup", "VAR62":"ia_cultura","VAR63":"ia_esporte", "VAR64":"ia_aeroporto", "VAR65":"ia_jornal", "VAR66":"ia_transporte", "VAR95":"num_leitos"}, inplace=True)
tb_regic.head()  
	
```

```python
tb_analise = tb_regic.merge(tb_municipio[['cod_mun','populacao_2021','area', 'densidade_2021']], how='left', left_on='cod_mun', right_on='cod_mun')
tb_analise['cod_mun_new'] = tb_analise['cod_mun'] // 10
tb_analise['CGT'] = tb_analise['CGT'].fillna(0)
tb_analise['CI'] = tb_analise['CI'].fillna(0)
tb_analise['CGP'] = tb_analise['CGP'].fillna(0)
tb_analise.head()
```

### Tabela de dados COVID

Tabela que apresenta os dados tratados do Covid

```python
tb_covid = pd.DataFrame(df_covid[["codmun", "data", "semanaEpi", "casosAcumulado", "obitosAcumulado"]])
tb_covid.dropna(subset=['codmun'], inplace=True)
tb_covid.head()
```

```python
#Convertendo para os tipos corretos
tb_covid['data'] = pd.to_datetime(tb_covid['data'])
tb_covid['codmun'] = tb_covid['codmun'].astype(int)
tb_covid['casosAcumulado'] = tb_covid['casosAcumulado'].astype(int)
```

```python
import datetime
data_1 = pd.Timestamp(2020, 4, 1)
data_2 = pd.Timestamp(2020, 5, 1)
data_3 = pd.Timestamp(2020, 6, 1)
data_4 = pd.Timestamp(2020, 7, 1)

tb_covid_1 = tb_covid[(tb_covid["data"] == data_1)]
tb_covid_2 = tb_covid[(tb_covid["data"] == data_2)]
tb_covid_3 = tb_covid[(tb_covid["data"] == data_3)]
tb_covid_4 = tb_covid[(tb_covid["data"] == data_4)]
```

Merge da tabela de analise e covid para realizar analises de correlação

```python
df_anal_1 = tb_analise.merge(tb_covid_1, how='left', left_on='cod_mun_new', right_on='codmun')
df_anal_2 = tb_analise.merge(tb_covid_2, how='left', left_on='cod_mun_new', right_on='codmun')
df_anal_3 = tb_analise.merge(tb_covid_2, how='left', left_on='cod_mun_new', right_on='codmun')
df_anal_4 = tb_analise.merge(tb_covid_4, how='left', left_on='cod_mun_new', right_on='codmun')
```

Medindo a correlacao das variáveis

```python
print(df_anal_4.corr(method='pearson'))
```
