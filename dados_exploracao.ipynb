## Verificando incoscistência na base Região de Saúde

```python
#Dados coletados por nós inicialmente
df_regiao_saude = pd.read_csv('data/regiao_saude_geo.csv', sep=",")
df_regiao_saude.columns = names=["cod_reg_saude", "nome_reg_saude", "latitude", "longitude", "limite_geografico"]
# df_regiao_saude.head()

#Dados coletados DataSus
df_saude = pd.read_csv('data/tb_regsaud.csv', sep=",", encoding='latin1')
# df_saude.head()
df_saude_ativo = df_saude[df_saude.CO_STATUS == "ATIVO"]
df_saude_ativo.shape

```

```python
df_merge = df_saude_ativo.merge(df_regiao_saude[['cod_reg_saude','nome_reg_saude']], how='outer', left_on='CO_REGSAUD', right_on='cod_reg_saude')
df_merge.shape
```

```python
df_merge[df_merge["cod_reg_saude"].isnull()]
```
