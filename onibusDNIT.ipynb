```python
import pandas as pd
df = pd.read_excel("PesquisaOD_2016_2017.xlsx", sheet_name=['BD_1PNT2016','BD_2PNT2016','BD_1PNT2017','BD_2PNT2017'])
#Separando abas do arquivo
df2 = df ['BD_1PNT2016']
df3 = df['BD_2PNT2016']
df4 = df['BD_1PNT2017']
df5= df['BD_2PNT2017']
```

```python
df_merged = pd.concat([df2, df3])
df_merged = df_merged.loc[:, [ 'POrigem', 'MOrigem', 'UFOrigem','GeocodOrigem', 'PDestino', 'MDestino', 'UFDestino','GeocodDestino', 'FreqPeriodo','FreqQnt','Classe', 'QtdPessoasVeic','considerar_parOD?','considerar_socio?']] 

df_merged2 = pd.concat([df4, df5])
df_merged2 = df_merged2.loc[:, [ 'POrigem', 'MOrigem', 'UFOrigem','GeocodOrigem', 'PDestino', 'MDestino', 'UFDestino','GeocodDestino', 'FreqPeriodo','FreqQnt','Classe', 'QtdPessoasVeic','considerar_parOD?','considerar_socio?']] 
#Retirando origens e destinos que são diferentes do Brasil
df_merged = df_merged.drop(df_merged[df_merged['PDestino'] != 'Brasil'].index)
df_merged = df_merged.drop(df_merged[df_merged['POrigem'] != 'Brasil'].index)
df_merged = df_merged.drop(df_merged[df_merged['FreqPeriodo'] == 'Informação não registrada'].index)
df_merged = df_merged.drop(df_merged[df_merged['QtdPessoasVeic'] == 'Informação não registrada'].index)
df_merged = df_merged.drop(df_merged[df_merged['considerar_parOD?'] == '0'].index)
df_merged = df_merged.drop(df_merged[df_merged['considerar_socio?'] == '0'].index)
df_merged = df_merged.dropna(subset=['QtdPessoasVeic'])
df_merged['FreqQnt'] = df_merged['FreqQnt'].replace('>10', 11)

#Retirando origens e destinos que são diferentes do Brasil
df_merged2 = df_merged2.drop(df_merged2[df_merged2['PDestino'] != 'Brasil'].index)
df_merged2 = df_merged2.drop(df_merged2[df_merged2['POrigem'] != 'Brasil'].index)
df_merged2 = df_merged2.drop(df_merged2[df_merged2['FreqPeriodo'] == 'Informação não registrada'].index)
df_merged2 = df_merged2.drop(df_merged2[df_merged2['QtdPessoasVeic'] == 'Informação não registrada'].index)
df_merged2 = df_merged2.drop(df_merged2[df_merged2['considerar_parOD?'] == '0'].index)
df_merged2 = df_merged2.drop(df_merged2[df_merged2['considerar_socio?'] == '0'].index)
df_merged2 = df_merged2.dropna(subset=['QtdPessoasVeic'])
df_merged2['FreqQnt'] = df_merged2['FreqQnt'].replace('>10', 11)

fatores = [ 'O1','O2','O3']

# Filtrar os valores da coluna com base nos transportes
geral2016 = df_merged[df_merged['Classe'].isin(fatores)]
geral2017 = df_merged2[df_merged2['Classe'].isin(fatores)]

display(geral2016)
display(geral2017)

```

```python
geral2016.to_csv('Onibus2016.csv', index=True)
geral2017.to_csv('Onibus2017.csv', index=True)
```

```python
coluna_alvo = "QtdPessoasVeic"  # Substitua pelo nome real da coluna

# Calcula a média dos valores na coluna
media = geral2016[coluna_alvo].mean()
media2 = geral2017[coluna_alvo].mean()
desvio_padrao = geral2016[coluna_alvo].std()
desvio_padrao2 = geral2017[coluna_alvo].std()

print("Média 2016:", media)
print("Média 2017:", media2)

print("Desvio Padrão 2016:", desvio_padrao)

print("Desvio Padrão 2017:", desvio_padrao2)
```
