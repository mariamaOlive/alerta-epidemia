```python
import pandas as pd


vendas_df = pd.read_csv("transporte_aereo_2019.csv",encoding='utf-8',sep=",")
municipios = pd.read_csv("municipio.csv",sep=",") 
display(vendas_df)
municipios = municipios.drop('latitude', axis=1)
municipios = municipios.drop('longitude', axis=1)
municipios = municipios.drop('populacao_2011', axis=1)
municipios = municipios.drop('populacao_2021', axis=1)
municipios = municipios.drop('area', axis=1)
municipios = municipios.drop('nome_uf', axis=1)
municipios = municipios.drop('cod_reg_saude', axis=1)
municipios = municipios.drop('densidade_2021', axis=1)
municipios = municipios.drop('cod_uf', axis=1)    
    
vendas_df = vendas_df.drop('EMPRESA_NOME', axis=1)
vendas_df = vendas_df.drop('Código OACI', axis=1)
vendas_df = vendas_df.drop('EMPRESA_NACIONALIDADE', axis=1)
vendas_df = vendas_df.drop('NATUREZA', axis=1)
vendas_df = vendas_df.drop('DECOLAGENS', axis=1)
vendas_df = vendas_df.drop('ASSENTOS', axis=1)
vendas_df = vendas_df.drop('AEROPORTO_DE_DESTINO_PAIS', axis=1)
vendas_df = vendas_df.drop('AEROPORTO_DE_DESTINO_SIGLA', axis=1)
vendas_df = vendas_df.drop('AEROPORTO_DE_ORIGEM_PAIS', axis=1)
vendas_df = vendas_df.drop('EMPRESA_SIGLA', axis=1)

vendas_df.columns = ['Cod','Ano','Mes','origem','UFOrigem','destino','UFDestino','PASSAGEIROS_PAGOS','PASSAGEIROS_GRATIS']

display(vendas_df)
```

```python
from unicodedata import normalize
def match_codigo(cidade):
    return (normalize('NFKD', cidade).encode('ASCII','ignore').decode('ASCII')).upper()

vendas_df["Origem"] = vendas_df.apply(lambda linha: match_codigo(linha.origem), axis= 1)
vendas_df["Destino"] = vendas_df.apply(lambda linha: match_codigo(linha.destino), axis= 1)
municipios["cidade"] = municipios.apply(lambda linha: match_codigo(linha.nome_mun), axis= 1)
municipios = municipios.drop('nome_mun', axis=1)
display(municipios)

vendas_df = vendas_df.drop('destino', axis=1)
vendas_df = vendas_df.drop('origem', axis=1)
vendas_df = vendas_df[['Cod','Ano','Mes','Origem','UFOrigem','Destino','UFDestino','PASSAGEIROS_PAGOS','PASSAGEIROS_GRATIS']]

display(vendas_df)
```

```python
left = vendas_df
right = municipios
df_final = pd.merge(
    left,
    right,
    how="left",
    left_on=['Origem','UFOrigem'],
    right_on=['cidade','uf'],    
    left_index=False,
    right_index=False,
    suffixes=("_l", "_r"),
)
display(df_final)

left = df_final
right = municipios
df_final1 = pd.merge(
    left,
    right,
    how="left",
    left_on=['Destino','UFDestino'],
    right_on=['cidade','uf'],    
    left_index=False,
    right_index=False,
    suffixes=("_l", "_r"),
)

df_final1 = df_final1.drop('uf_l', axis=1)
df_final1 = df_final1.drop('pais_l', axis=1)
df_final1 = df_final1.drop('cidade_l', axis=1)
df_final1 = df_final1.drop('uf_r', axis=1)
df_final1 = df_final1.drop('pais_r', axis=1)
df_final1 = df_final1.drop('cidade_r', axis=1)

df_final1.columns = ['Cod','Ano','Mes','Origem','UFOrigem','Destino','UFDestino','PASSAGEIROS_PAGOS','PASSAGEIROS_GRATIS','cod_mun_origem','cod_mun_destino']
df_final1 = df_final1[['Cod','Mes','Ano','Origem','UFOrigem','cod_mun_origem','Destino','UFDestino','cod_mun_destino','PASSAGEIROS_PAGOS','PASSAGEIROS_GRATIS']]

df_final1["Origem"] = df_final1["Origem"] +"/"+ df_final1["UFOrigem"]
df_final1["Destino"] = df_final1["Destino"] +"/"+ df_final1["UFDestino"]
df_final1['passageiros'] = df_final1['PASSAGEIROS_PAGOS'] + df_final1['PASSAGEIROS_GRATIS']
df_final1 = df_final1.drop('UFOrigem', axis=1)
df_final1 = df_final1.drop('UFDestino', axis=1)
df_final1 = df_final1.drop('PASSAGEIROS_PAGOS', axis=1)
df_final1 = df_final1.drop('PASSAGEIROS_GRATIS', axis=1)

df = df_final1.groupby(['Origem','Destino','Mes']).agg({'passageiros':'sum','cod_mun_origem':'max','cod_mun_destino':'max','Ano' :'max'})


display(df)
df.to_csv("aeriosFinal.csv")
```
