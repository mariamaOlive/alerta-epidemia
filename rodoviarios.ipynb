```python
import pandas as pd
import os
import glob

path = os.getcwd()
csv_files = glob.glob(os.path.join("/Users/lucas/Desktop/ufrpe/Untitled Folder/passagens", "*.csv"))
vendas_df=pd.DataFrame()

for f in csv_files:
    df = pd.read_csv(f,encoding='latin1',sep=";")
    vendas_df = pd.concat([vendas_df, df], axis=0)
    
#vendas_df = pd.read_csv("venda_passagem_02_2019.csv",encoding='latin1',sep=";")
municipios = pd.read_csv("municipio.csv",sep=",") 

display(vendas_df)

vendas_df.columns = ['MesEmissão','MesViagem','Origem','Destino','UFOrigem','UFDestino','mediavalor','dpvalototal','QuantidaDeBilhetes']


vendas_df = vendas_df.drop('mediavalor', axis=1)
vendas_df = vendas_df.drop('dpvalototal', axis=1)

municipios = municipios.drop('latitude', axis=1)
municipios = municipios.drop('longitude', axis=1)
municipios = municipios.drop('populacao_2011', axis=1)
municipios = municipios.drop('populacao_2021', axis=1)
municipios = municipios.drop('area', axis=1)
municipios = municipios.drop('nome_uf', axis=1)
municipios = municipios.drop('cod_reg_saude', axis=1)
municipios = municipios.drop('densidade_2021', axis=1)
municipios = municipios.drop('cod_uf', axis=1)

vendas_df["Origem"],vendas_df["UFOrigem"] = vendas_df["Origem"].str.split('/').str
vendas_df["Destino"],vendas_df["UFDestino"]= vendas_df["Destino"].str.split('/').str

vendas_df = vendas_df[['MesEmissão','MesViagem','Origem','UFOrigem','Destino','UFDestino','QuantidaDeBilhetes']]

display(vendas_df)
display(municipios)
```

```python
from unicodedata import normalize
def match_codigo(cidade):
    return (normalize('NFKD', cidade).encode('ASCII','ignore').decode('ASCII')).upper()

municipios["cidade"] = municipios.apply(lambda linha: match_codigo(linha.nome_mun), axis= 1)
display(municipios)
```

```python
left = vendas_df
right = municipios
df_final = pd.merge(
    left,
    right,
    how="left",
    left_on=['Origem','UFOrigem'],
    right_on=['cidade','uf'],    
    left_index=False,
    right_index=False,
    suffixes=("_l", "_r"),
)
display(df_final)

left = df_final
right = municipios
df_final1 = pd.merge(
    left,
    right,
    how="left",
    left_on=['Destino','UFDestino'],
    right_on=['cidade','uf'],    
    left_index=False,
    right_index=False,
    suffixes=("_l", "_r"),
)

display(df_final1)

```

```python
df_final1 = df_final1.drop('nome_mun_l', axis=1)
df_final1 = df_final1.drop('uf_l', axis=1)
df_final1 = df_final1.drop('pais_l', axis=1)
df_final1 = df_final1.drop('cidade_l', axis=1)
df_final1 = df_final1.drop('nome_mun_r', axis=1)
df_final1 = df_final1.drop('uf_r', axis=1)
df_final1 = df_final1.drop('pais_r', axis=1)
df_final1 = df_final1.drop('MesEmissão', axis=1)
df_final1 = df_final1.drop('cidade_r', axis=1)
df_final1["origem"] = df_final1["Origem"] +"/"+ df_final1["UFOrigem"]
df_final1["destino"] = df_final1["Destino"] +"/"+ df_final1["UFDestino"]
df_final1 = df_final1.drop('Origem', axis=1)
df_final1 = df_final1.drop('Destino', axis=1)
df_final1 = df_final1.drop('UFOrigem', axis=1)
df_final1 = df_final1.drop('UFDestino', axis=1)
display(df_final1)
df = df_final1.groupby(['origem','destino','MesViagem']).agg({'QuantidaDeBilhetes':'sum','cod_mun_l':'max','cod_mun_r':'max'})

display(df)
df.to_csv("rodoviariosFinal.csv")
```
