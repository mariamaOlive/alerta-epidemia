```python
import pandas as pd

# DataFrame de exemplo
data = {'Nome': ['João', 'Maria', 'João', 'Maria'],
        'Valor1': [10, 20, 30, 40],
        'Valor2': [5, 10, 15, 20],
        'Valor3': [2, 4, 6, 8]}
df = pd.DataFrame(data)
print(df)

# Definindo as colunas para soma e as colunas para manter iguais
columns_to_sum = ['Valor1', 'Valor3']
columns_to_keep = ['Nome', 'Valor2']

# Agrupando e aplicando operações de agregação nas colunas desejadas
df_grouped = df.groupby('Nome').agg({'Valor2': 'first', **{col: 'sum' for col in columns_to_sum}})

print(df_grouped)
```
