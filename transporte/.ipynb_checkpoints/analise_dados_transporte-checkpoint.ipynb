Distribuição DNIT

Processando dados do DNIT para o mesmo formato do IBGE

```python
def processar_dnit(df):

    #Removendo campos nulos
    df_dnit = df[(df['GeocodOrigem'].notna()) & (df['GeocodDestino'].notna())]
    df_dnit = df_dnit.astype({"GeocodOrigem": int, "GeocodDestino": int})

    # #Filtrando observações por tipo de automóvel
    df_dnit = df_dnit[(df_dnit['Classe']=="P2") 
             | (df_dnit['Classe']=="O1") 
             | (df_dnit['Classe']=="O2") 
             | (df_dnit['Classe']=="O3")] 
    
    df_dnit = pd.DataFrame({'freq_semanal' : df_dnit.groupby( [ "GeocodOrigem", "GeocodDestino" ] ).size()}).reset_index()
    
    #Removendo ligações que possuem mesma origem e destino
    df_dnit = df_dnit[df_dnit["GeocodOrigem"]!=df_dnit["GeocodDestino"]]
    return df_dnit 
```

```python
# Processando dados do DNIT
lista_df_dnit_proc = []
for df in lista_df_dnit:
    lista_df_dnit_proc.append(processar_dnit(df))
```

Distribuição de Origem e Destino

```python
df_estado = pd.read_csv("data/estado.csv")
```

```python
def adicionar_estado(df_estado, df_dnit):
    df_dnit["estado_origem"] = df_dnit["GeocodOrigem"].astype(str).str[:2]
    df_dnit["estado_origem"] = df_dnit["estado_origem"].astype(int)
    df_dnit["estado_destino"] = df_dnit["GeocodDestino"].astype(str).str[:2]
    df_dnit["estado_destino"] = df_dnit["estado_destino"].astype(int)

    df_dnit = df_dnit.merge(df_estado[["cod_uf", "uf"]], left_on="estado_origem", right_on= "cod_uf")
    df_dnit.rename(columns={"uf": "uf_origem"}, inplace=True)
    df_dnit = df_dnit.merge(df_estado[["cod_uf", "uf"]], left_on="estado_destino", right_on= "cod_uf")
    df_dnit.rename(columns={"uf": "uf_destino"}, inplace=True)

    return df_dnit
```

```python
lista_temp_df_dnit_proc = []
for df in lista_df_dnit_proc:
    lista_temp_df_dnit_proc.append(adicionar_estado(df_estado, df))
lista_df_dnit_proc = lista_temp_df_dnit_proc
```

## Comparando dados do DNIT com IBGE

Consertando dados para fazer o merge

```python
## Duplicando dados do IBGE para que as ligações tenham os 2 sentidos

df_ibge_1 = df_ibge[["CODMUNDV_A", "CODMUNDV_B","VAR06"]].copy(deep=True)
df_ibge_2 = df_ibge[["CODMUNDV_A", "CODMUNDV_B","VAR06"]].copy(deep=True)
df_ibge_2 = df_ibge_2.rename(columns={"CODMUNDV_B": "CODMUNDV_A", "CODMUNDV_A": "CODMUNDV_B"})
df_ibge_final = pd.concat([df_ibge_1, df_ibge_2])
df_ibge_final[df_ibge_final.duplicated()]

## Removendo valores 0 de frequência
df_ibge_final = df_ibge_final[df_ibge_final["VAR06"]>0]
df_ibge_final.shape

## Renomeando colunas do df do IBGE e DNIT
#IBGE
df_ibge_final = df_ibge_final.rename(columns={"VAR06": "ibge", "CODMUNDV_A": "origem", "CODMUNDV_B": "destino"})
#DNIT
for df_dnit in lista_df_dnit_proc:
    df_dnit.rename(columns={"freq_semanal": "dnit", "GeocodOrigem": "origem", "GeocodDestino": "destino"}, inplace=True)
```

OBS.:
Após duplicar os dados para conter os dois sentidos ficamos com 130528 ligações no dados do IBGE.

```python
#Realiza o merge entre os dataframes
def merge_dnit_ibge(df_dnit, df_ibge):
    df = df_ibge.merge(df_dnit, left_on=['origem','destino'], right_on = ['origem','destino'])

    columns_titles = ["origem","destino", "ibge", "dnit"]
    df=df.reindex(columns=columns_titles)
    return df
```

```python
lista_df_merge = []
for df_dnit in lista_df_dnit_proc:
    lista_df_merge.append(merge_dnit_ibge(df_dnit, df_ibge_final))
```

Verificando se os dados possuem a mesma tendência

```python
for nome, df in zip(nomes_planilhas, lista_df_merge):
    print(nome)
    print(df[["ibge","dnit"]].corr())
    print()
```

Plotando dados no scatterplot

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 11))
for df, nome,ax in zip(lista_df_merge, nomes_planilhas ,axes.flat):
    sns.scatterplot(data=df, x="ibge", y="dnit", ax=ax).set(title=nome)
plt.show()
```

## Avaliando ligações dentro de um *Arranjo Populacional*

```python
# Carregando df de municipio para análise
df_municipios = pd.read_csv("data/municipio.csv")

# Adiciona o codigo do arranjo no município IBGE e DNIT
def adiciona_cod_arranjo(df_municipios, df):
    df = df.merge(df_municipios[['cod_mun','cod_arranjo']], left_on=['origem'], right_on = ['cod_mun'])
    df = df.drop(columns=['cod_mun'])
    df.rename(columns={"cod_arranjo": "cod_arranjo_origem"}, inplace=True)
    df = df.merge(df_municipios[['cod_mun','cod_arranjo']], left_on=['destino'], right_on = ['cod_mun'])
    df.rename(columns={"cod_arranjo": "cod_arranjo_destino"}, inplace=True)
    df = df.drop(columns=['cod_mun'])
    return df
```

```python
# Realizando merge com info de municipios nas ligações do IBGE e DNIT

# IBGE
df_ibge_final = adiciona_cod_arranjo(df_municipios, df_ibge_final)

# DNIT
lista_temp = []
for df_dnit in lista_df_dnit_proc:
    lista_temp.append(adiciona_cod_arranjo(df_municipios, df_dnit))
lista_df_dnit_proc = lista_temp
```

### Análise ligações IBGE

```python
# Verificar quantas ligações fazem parte do mesmo Arranjo Populacional

df_mesmo_ibge = df_ibge_final[df_ibge_final['cod_arranjo_origem']==df_ibge_final['cod_arranjo_destino']]
n_df = df_ibge_final.shape[0]
n_mesmo_arr = df_mesmo_ibge.shape[0]

print("Análise do IBGE")
print("Total de ligações: " + str(n_df))
print("Total de ligações no mesmo arranjo: " + str(n_mesmo_arr))
print("Porcentagem de ligações no mesmo arranjo: " + str(round(n_mesmo_arr * 100/ n_df, 2)))
print()
```

### Análise ligações DNIT

```python
for nome, df in zip(nomes_planilhas, lista_df_dnit_proc):
    df_mesmo_arr = df[df['cod_arranjo_origem']==df['cod_arranjo_destino']]
    n_mesmo_arr = df_mesmo_arr.shape[0]
    n_df = df.shape[0]

    print("Análise de " + nome)
    print("Total de ligações: " + str(n_df))
    print("Total de ligações no mesmo arranjo: " + str(n_mesmo_arr))
    print("Porcentagem de ligações no mesmo arranjo: " + str(round(n_mesmo_arr * 100/ n_df, 2)))
    print()
```
