```python
import pandas as pd

df= pd.read_excel("../Base_de_dados_ligacoes_rodoviarias_e_hidroviarias_2016.xlsx", sheet_name=['Base de dados'])
df2 = df['Base de dados']
```

```python
dfIGBE = df2.rename(columns={'CODMUNDV_A': 'GeocodOrigem', 'CODMUNDV_B': 'GeocodDestino','VAR05':'PessoasHidro','VAR06':'PessoasRod'})
dfIGBE = dfIGBE.loc[dfIGBE['UF_A'].str.len()<3]
dfIGBE = dfIGBE.loc[dfIGBE['UF_B'].str.len()<3]
dfIGBE['PessoasAnoR'] = dfIGBE['PessoasRod'] * 22 * 52
dfIGBE['PessoasAnoH'] = dfIGBE['PessoasHidro'] * 22 * 52

display(dfIGBE)
```

```python
IbgeRod = dfIGBE.loc[:, [ 'GeocodOrigem','UF_A',  'NOMEMUN_A','GeocodDestino', 'UF_B', 'NOMEMUN_B','PessoasRod','PessoasAnoR']]
IbgeHidro = dfIGBE.loc[:, [ 'GeocodOrigem', 'UF_A','NOMEMUN_A', 'GeocodDestino','UF_B', 'NOMEMUN_B','PessoasHidro','PessoasAnoH']]

IbgeRod = IbgeRod.query('PessoasRod != 0')
IbgeHidro = IbgeHidro.query('PessoasHidro != 0')

IbgeRod2 = IbgeRod.copy()

# Renomeie as colunas de origem para destino e vice-versa
IbgeRod2.rename(columns={
    'GeocodOrigem': 'GeocodDestino',
    'UF_A': 'UF_B',
    'NOMEMUN_A': 'NOMEMUN_B',
    'GeocodDestino': 'GeocodOrigem',
    'UF_B': 'UF_A',
    'NOMEMUN_B': 'NOMEMUN_A'
}, inplace=True)

# Concatene o DataFrame original e o DataFrame modificado
IbgeRod = pd.concat([IbgeRod, IbgeRod2], ignore_index=True)

IbgeHidro2 = IbgeHidro.copy()

# Renomeie as colunas de origem para destino e vice-versa
IbgeHidro2.rename(columns={
    'GeocodOrigem': 'GeocodDestino',
    'UF_A': 'UF_B',
    'NOMEMUN_A': 'NOMEMUN_B',
    'GeocodDestino': 'GeocodOrigem',
    'UF_B': 'UF_A',
    'NOMEMUN_B': 'NOMEMUN_A'
}, inplace=True)

# Concatene o DataFrame original e o DataFrame modificado
IbgeHidro = pd.concat([IbgeHidro, IbgeHidro2], ignore_index=True)


display(IbgeRod)
display(IbgeHidro)
```

Duplicar cada linha invertendo origem e destino com mesmo valor de frequencia

```python
IbgeRod.to_csv('IBGE_R.csv', index=True)
IbgeHidro.to_csv('IBGE_H.csv', index=True)
```
