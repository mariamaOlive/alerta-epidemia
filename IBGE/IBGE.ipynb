```python
import pandas as pd

df= pd.read_excel("../Base_de_dados_ligacoes_rodoviarias_e_hidroviarias_2016.xlsx", sheet_name=['Base de dados'])
df2 = df['Base de dados']
```

```python
dfIGBE = df2.rename(columns={'CODMUNDV_A': 'GeocodOrigem', 'CODMUNDV_B': 'GeocodDestino','VAR05':'PessoasHidro','VAR06':'PessoasRod'})
dfIGBE = dfIGBE.loc[dfIGBE['UF_A'].str.len()<3]
dfIGBE = dfIGBE.loc[dfIGBE['UF_B'].str.len()<3]
dfIGBE['PessoasAnoR'] = dfIGBE['PessoasRod'] * 22 * 52
dfIGBE['PessoasAnoH'] = dfIGBE['PessoasHidro'] * 22 * 52

display(dfIGBE)
```

```python
IbgeRod = dfIGBE.loc[:, [ 'GeocodOrigem','UF_A',  'NOMEMUN_A','GeocodDestino', 'UF_B', 'NOMEMUN_B','PessoasRod','PessoasAnoR']]
IbgeHidro = dfIGBE.loc[:, [ 'GeocodOrigem', 'UF_A','NOMEMUN_A', 'GeocodDestino','UF_B', 'NOMEMUN_B','PessoasHidro','PessoasAnoH']]

IbgeRod = IbgeRod.query('PessoasRod != 0')
IbgeHidro = IbgeHidro.query('PessoasHidro != 0')

IbgeRod2 = IbgeRod.copy()

# Renomeie as colunas de origem para destino e vice-versa
IbgeRod2.rename(columns={
    'GeocodOrigem': 'GeocodDestino',
    'UF_A': 'UF_B',
    'NOMEMUN_A': 'NOMEMUN_B',
    'GeocodDestino': 'GeocodOrigem',
    'UF_B': 'UF_A',
    'NOMEMUN_B': 'NOMEMUN_A'
}, inplace=True)

# Concatene o DataFrame original e o DataFrame modificado
IbgeRod = pd.concat([IbgeRod, IbgeRod2], ignore_index=True)

IbgeHidro2 = IbgeHidro.copy()

# Renomeie as colunas de origem para destino e vice-versa
IbgeHidro2.rename(columns={
    'GeocodOrigem': 'GeocodDestino',
    'UF_A': 'UF_B',
    'NOMEMUN_A': 'NOMEMUN_B',
    'GeocodDestino': 'GeocodOrigem',
    'UF_B': 'UF_A',
    'NOMEMUN_B': 'NOMEMUN_A'
}, inplace=True)

# Concatene o DataFrame original e o DataFrame modificado
IbgeHidro = pd.concat([IbgeHidro, IbgeHidro2], ignore_index=True)


display(IbgeRod)
display(IbgeHidro)
```

Duplicar cada linha invertendo origem e destino com mesmo valor de frequencia

```python
IbgeRod.to_csv('IBGE_R.csv', index=True)
IbgeHidro.to_csv('IBGE_H.csv', index=True)
```

```python
DF_IBGE = pd.concat([IbgeHidro, IbgeRod], ignore_index=False)
DF_IBGE['PessoasAnoH'].fillna(0, inplace=True)
DF_IBGE['PessoasAnoR'].fillna(0, inplace=True)
DF_IBGE['PessoasAno'] = DF_IBGE['PessoasAnoH'] + DF_IBGE['PessoasAnoR']
display(DF_IBGE)

contagem_coluna1 = DF_IBGE['UF_A'].value_counts()
print("Contagem de Abrangencia de ORIGEM':\n", contagem_coluna1)

contagem_coluna2 = DF_IBGE['UF_B'].value_counts()
print("Contagem de Abrangencia de DESTINO':\n", contagem_coluna2)
```

```python
# Mapeamento de estados para regiões
mapa_regioes = {
    'SP': 'Sudeste',
    'RJ': 'Sudeste',
    'MG': 'Sudeste',
    'ES': 'Sudeste',
    'PR': 'Sul',
    'SC': 'Sul',
    'RS': 'Sul',
    'MS': 'Centro-Oeste',
    'MT': 'Centro-Oeste',
    'GO': 'Centro-Oeste',
    'DF': 'Centro-Oeste',
    'BA': 'Nordeste',
    'SE': 'Nordeste',
    'AL': 'Nordeste',
    'PE': 'Nordeste',
    'PB': 'Nordeste',
    'RN': 'Nordeste',
    'CE': 'Nordeste',
    'PI': 'Nordeste',
    'MA': 'Nordeste',
    'PA': 'Norte',
    'AM': 'Norte',
    'RR': 'Norte',
    'AP': 'Norte',
    'TO': 'Norte',
    'RO': 'Norte',
    'AC': 'Norte'
}

# Criando uma nova coluna para a região
DF_IBGE['regiao'] = DF_IBGE['UF_A'].map(mapa_regioes)

# Agrupando por região e somando os valores
resultado = DF_IBGE.groupby('regiao')['PessoasAno'].sum()

resultado_formatado = resultado.apply(lambda x: "{:,.2f}".format(x).replace(',', 'X').replace('.', ',').replace('X', '.'))

print("Resultados formatados:\n", resultado_formatado)
```

```python
# Contando quantas vezes cada estado aparece dentro de cada região
contagem_regioes = DF_IBGE.groupby('regiao')['UF_A'].value_counts()

contagem_total_regioes = DF_IBGE.groupby('regiao').size()

# Convertendo a contagem 
contagem_total_regioes_formatado = contagem_total_regioes.apply(lambda x: "{:,.0f}".format(x).replace(',', 'X').replace('.', ',').replace('X', '.'))

#print("Contagem de estados por região:\n", contagem_regioes)
print("\nContagem total por região, formatada:\n", contagem_total_regioes_formatado)
```
